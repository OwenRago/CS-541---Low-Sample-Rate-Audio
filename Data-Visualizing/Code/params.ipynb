```python
import numpy as np
import matplotlib.pyplot as plt

df = np.load('../Data/params.npy')
```

```python

params = np.mean(df, axis=1) /1000000
print(params)
plt.figure(figsize=(10, 6))

plt.plot(params,'o-')

plt.title('Parameters by Model Size')
plt.xlabel('Model Size')
plt.ylabel('Parameters (10^6)')
plt.xticks(np.arange(len(params)), range(len(params)))
plt.grid()
plt.savefig('../Images/Params.png')
plt.show()

```
