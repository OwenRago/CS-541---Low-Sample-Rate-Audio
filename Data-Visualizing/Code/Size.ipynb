```python
import numpy as np
import matplotlib.pyplot as plt

df = np.load('../Data/Model_size_data.npy')
df = np.swapaxes(df, 0, 1)
```

```python
KHZ = ['2', '4', '8', '16', '22']
# plt.figure(figsize=(10, 6))

for i in range(df.shape[1]):
    for j in range(df.shape[0]):
        plt.plot(df[j, i],'o-', label=f'{KHZ[j]} kHz')

    plt.title(f'Accuracy of model size {i+1} at each Epoch')
    plt.xlabel('Epochs')
    plt.ylabel('Accuracy')
    plt.legend()
    plt.grid()
    plt.savefig(f"../Images/Acc_Epoch_Size{i}.png")
    plt.show()
```

```python
plt.figure(figsize=(10, 6))
max_accuracies = np.max(df, axis=2)

x = np.arange(len(KHZ))

bar_width = 0.2

for j in range(max_accuracies.shape[1]):
    plt.bar(x + (j - 1) * bar_width, max_accuracies[:, j], width=bar_width, label=f'Parameters {j+1}')

plt.xlabel('Frequency (kHz)')
plt.ylabel('Maximum Accuracy')
plt.title('Maximum Accuracy by Model Size and Dataset')
plt.xticks(x, KHZ)
plt.ylim(0.6, 1.0)
plt.legend()
plt.grid(axis='y', linestyle='-', alpha=0.4)
plt.savefig('../Images/Acc_Model.png')
plt.show()
```
