```python
import numpy as np
import matplotlib.pyplot as plt

df = np.load('../Data/macs.npy')
```

```python

macs = np.mean(df, axis=1) / 1000000000
print(macs)
plt.figure(figsize=(10, 6))

plt.plot(macs,'o-')

plt.title('MACs by Model Size')
plt.xlabel('Model Size')
plt.ylabel('MACs (10^9)')
plt.xticks(np.arange(len(macs)), range(len(macs)))
plt.grid()
plt.savefig('../Images/MACs.png')
plt.show()

```
