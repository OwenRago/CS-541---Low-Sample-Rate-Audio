```python
import numpy as np
import matplotlib.pyplot as plt

HZ = [2, 4, 8, 16, 22]
df = np.load('../Data/LSTM_params.npy')
print(df)
```

```python
params = df/1000000
print(params)

plt.figure(figsize=(10, 6))

for i in range(params.shape[0]):
    plt.plot(params[i], 'o-', label=f'{HZ[i]} kHz')

plt.title('Parameters by Model Size across Different Frequencies')
plt.xlabel('Model Size')
plt.ylabel('Parameters (10^6)')
plt.xticks(range(params.shape[1]))
plt.legend()
plt.grid()
plt.yscale('log')
plt.savefig('../Images/LSTM_Params.png')
plt.show()

```
