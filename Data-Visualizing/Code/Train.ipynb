```python
import numpy as np
import matplotlib.pyplot as plt

df = np.load('../Data/Model_train_data.npy')
df = np.swapaxes(df, 0, 1)
```

```python
KHZ = ['2', '4', '8', '16', '22']
plt.figure(figsize=(10, 6))
average_time = np.sum(df, axis=2) / df.shape[2]
x = np.arange(len(KHZ))

bar_width = 0.2

for i in range(average_time.shape[1]):
    plt.bar(x + (i - 1) * bar_width, average_time[:, i], width=bar_width, label=f'Parameters {i+1}')

plt.xlabel('Frequency (kHz)')
plt.ylabel('Average Training Time (s)')
plt.title('Average Training Time by Model Size and Dataset')
plt.xticks(x, KHZ)
plt.legend()
plt.grid(axis='y', linestyle='-', alpha=0.4)
plt.savefig('../Images/Train_Model.png')
plt.show()
```
